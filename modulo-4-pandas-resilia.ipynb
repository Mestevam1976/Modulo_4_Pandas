<font color="#000099">**Região:**</font>
Variável categórica nominal. 
É a sigla que representa a região do país, podendo ser N para Norte, NE para Nordeste, CO para Centro-oeste, S para Sul, SE para Sudeste. 

<font color="#000099">**Estado:**</font>
Variável categórica nominal. 
É a sigla que representa o [estado](https://duvidas.dicio.com.br/siglas-dos-estados-brasileiros/). 

<font color="#000099">**Município:**</font>
Variável categórica nominal. 
Nome do município onde a revenda ocorreu.

<font color="#000099">**Revenda:**</font>
Variável categórica nominal. 
Empresa que realizou a revenda do produto. 

<font color="#000099">**CNPJ da Revenda:**</font>
Variável categórica nominal. 
Número do Cadastro Nacional da Pessoa Jurídica da empresa que realizou a revenda. 
Formato 00.000.000/0000-00 que possui o número de inscrição, matriz e dígitos verificadores separados por ponto, barra e traço.

<font color="#000099">**Nome da Rua:**</font>
Variável categórica nominal. 
Nome do logradouro. 

<font color="#000099">**Número Rua:**</font>
Variável categórica nominal. 
Numeração do prédio onde a venda ocorreu. 

<font color="#000099">**Complemento:**</font>
Variável categórica nominal. Dado complementar ao endereço, utilizado para indicar com exatidão a localização do imóvel.

<font color="#000099">**Bairro:**</font>
Variável categórica nominal. Nome do bairro/vila.

<font color="#000099">**Cep:**</font>
Variável categórica nominal. 
[Código de Endereçamento Postal](https://pt.wikipedia.org/wiki/C%C3%B3digo_de_Endere%C3%A7amento_Postal#Componentes). 
Formato 00000-000.

<font color="#000099">**Produto:**</font>
Variável categórica nominal. 
Indica o combustível comercializado.

<font color="#000099">**Data da Coleta:**</font>
Variável categórica nominal.
Data da coleta do dado. 
Formato original *dd*/*mm*/*yyyy*. 
Formato ISO *yyyy*-*mm*-*dd*.

<font color="#000099">**Valor de Venda:**</font>
Variável numérica contínua. 
Valor pago correspondente a variável "Unidade de Medida".

<font color="#000099">**Valor de Compra:**</font>
Esta variável não possui nenhuma entrada válida.

<font color="#000099">**Unidade de Medida:**</font>
Variável categórica nominal. 
Valor para qual a variável "Valor de Venda" se refere. 
Formato moeda / medida.

<font color="#000099">**Bandeira:**</font>
Variável categórica nominal. 
Indica vínculo comercial com grande marca ou simplesmente o valor "BRANCA" caso não seja um posto bandeirado.

# <h2><b>Funções para criar quinzenas e semanas<b><h2>

```python
def cria_quinzenas(dataframe):
    primeiro = (f"{dataframe['Data da Coleta'][0].year}-{dataframe['Data da Coleta'][0].month}-01")
    primeiro=pd.to_datetime(primeiro, format = "%Y/%m/%d")
    ultimo =(f"{dataframe['Data da Coleta'][len(dataframe)-1].year}-{dataframe['Data da Coleta'][len(dataframe)-1].month}-{dataframe['Data da Coleta'][len(dataframe)-1].day}")
    ultimo=pd.to_datetime(ultimo, format = "%Y/%m/%d")

    dic = {}
    total_dias = ultimo-primeiro
    quinzenas = total_dias/15
    dia=primeiro
    for i in range(1,quinzenas.days+1):
        dia_calc = (dia+timedelta(days=15))
        dic[i]=(dia_calc)
        dia = dia_calc
    dic[i+1]=ultimo
    
    dia=primeiro
    saida = pd.DataFrame()
    for i in dic.items():
        while dia <=i[1]:
            a = pd.DataFrame(data=[i[0]],index=[dia],columns=['Quinzena'])
            saida = pd.concat([saida,a])
            dia = dia + timedelta(days=1)
    df = pd.merge(right= dataframe,left = saida,how='right',right_on = ['Data da Coleta'],left_index=True)
    return df

def cria_semana(dataframe):
    primeiro = (f"{dataframe['Data da Coleta'][0].year}-{dataframe['Data da Coleta'][0].month}-01")
    primeiro=pd.to_datetime(primeiro, format = "%Y/%m/%d")
    ultimo =(f"{dataframe['Data da Coleta'][len(dataframe)-1].year}-{dataframe['Data da Coleta'][len(dataframe)-1].month}-{dataframe['Data da Coleta'][len(dataframe)-1].day}")
    ultimo=pd.to_datetime(ultimo, format = "%Y/%m/%d")

    dic = {}
    total_dias = ultimo-primeiro
    semanas = total_dias/7
    dia=primeiro
    for i in range(1,semanas.days+1):
        dia_calc = (dia+timedelta(days=7))
        dic[i]=(dia_calc)
        dia = dia_calc
    dic[i+1]=ultimo
    
    dia=primeiro
    saida = pd.DataFrame()
    for i in dic.items():
        while dia <=i[1]:
            a = pd.DataFrame(data=[i[0]],index=[dia],columns=['Semana'])
            saida = pd.concat([saida,a])
            dia = dia + timedelta(days=1)
    df = pd.merge(right= dataframe,left = saida,how='right',right_on = ['Data da Coleta'],left_index=True)
    return df
```

# <h2><b>Tratando os Dados</b><h2>

 <h3><b>Aqui observamos algumas particularidades dos dados, e decidimos fazer as seguintes alterações:</b></h3>
 <h3> * Como foi observado, a coluna <b>*'Valor de Compra'*</b> esta sem informações e por isso foi excluida.</h3>
 <h3> * A coluna <b>*'Valor de Venda'* </b>está como texto e foi transformada para o formato numerico float. </h3>
 <h3> * A coluna <b>*'Data da Coleta'*</b> estava no formato de texto e foi transformada para o formato Datetime.</h3> 

```python
# Excluindo a coluna 'Valor de venda que esta vazia'

dados_jun_jul.drop('Valor de Compra',inplace=True,axis=1)
dados_ano_2021.drop('Valor de Compra',inplace=True,axis=1)

# Convertendo coluna 'Valor de Venda' para float

jun["Valor de Venda"] = jun["Valor de Venda"].replace(",",".", regex=True).astype(float)
jul["Valor de Venda"] = jul["Valor de Venda"].replace(",",".", regex=True).astype(float)
dados_jun_jul["Valor de Venda"] = dados_jun_jul["Valor de Venda"].replace(",",".", regex=True).astype(float)
dados_ano_2021["Valor de Venda"] = dados_ano_2021["Valor de Venda"].replace(",",".", regex=True).astype(float)

#Convertendo coluna 'Data da Coleta' para date time

dados_jun_jul['Data da Coleta'] = pd.to_datetime(dados_jun_jul['Data da Coleta'],format = "%d/%m/%Y")
dados_ano_2021['Data da Coleta'] = pd.to_datetime(dados_ano_2021['Data da Coleta'],format = "%d/%m/%Y")
```

<h3><b>* Renomeamos algumas colunas para ficar mais fácil seu entendimento e uso no estudo</b></h3>

```python
dados_jun_jul.columns = ['Região', 'Estado', 'Município', 'Revenda',
       'CNPJ da Revenda', 'Nome da Rua', 'Número Rua', 'Complemento', 'Bairro',
       'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
       'Unidade de Medida', 'Bandeira']

dados_ano_2021.columns =['Região', 'Estado', 'Município', 'Revenda',
       'CNPJ da Revenda', 'Nome da Rua', 'Número Rua', 'Complemento', 'Bairro',
       'Cep', 'Produto', 'Data da Coleta', 'Valor de Venda',
       'Unidade de Medida', 'Bandeira']
```

<h2><b>Correlação</b></h2>

<h3><b>O Dataframe da Petrobrás possuía, originalmente, apenas uma coluna preenchida com valores numéricos e quantitativos e, por essa razão, não era possível estabelecer uma correlação probabilística com os dados originais.</b></h3>
<h3><b>Diante disso, tentamos criar as correlações entre as variantes do preço de venda, sendo elas: "Preço Mínimo", "Preço Médio" e "Preço Máximo" para saber se o fator regional impactava nos resultados</b></h3>

```python
# Para isso criamos um novo DataFrame com os valores de mínimo, médio e máximo:

maximas_por_região = dados_jun_jul.groupby(['Produto','Região'])[['Valor de Venda']].max()
maximas_por_região.columns = ['Máxima']
minimas_por_região = dados_jun_jul.groupby(['Produto','Região'])[['Valor de Venda']].min()
minimas_por_região.columns = ['Minima']
media_geral = pd.concat([media_por_regiao,maximas_por_região,minimas_por_região],axis=1)
media_geral.rename(columns={'Valor de Venda': 'Média'}).round(3)
```

```python
# Aplicando a correlação sobre o DataFrame criado:

media_geral.corr()
```

```python
# Plotando um gráfico de calor:

plt.figure(figsize=(10,5))
sns.heatmap(media_geral.corr(), vmin=-1.0, vmax=1.0, cmap='coolwarm', annot=True)
print()
```

<h3><b>O gráfico de calor acima indica uma forte correlação entre as variantes de preço, porém, não destaca a intensidade dessa correlação em função da região. Motivo pelo qual, testamos a existência de correlação através da Transposta do DataFrame</b></h3>

```python
# Aplicando a correlação com a visualização do Dataframe de forma Transposta:

plt.figure(figsize=(15,10))
sns.heatmap(media_geral.T.corr(), vmin=-1.0, vmax=1.0, cmap='coolwarm', annot=True)
print()
```

<h3><b>O Gráfico de calor resultante da Transposta do Dataframe de Variação de Preços, reitera a forte correlação entre as variáveis, independente da região amostrada, evidenciando com isso não ser um recurso válido para tomada de decisões nessa situação.</b></h3>
