## <b>Criando as Funções<b>
    
<h3> Aqui criam-se todas as funções que serão úteis para a análise. <h3>

**<font color="#000099">cria_quinzenas:</font>**

* Entrada: *DataFrame* com coluna do tipo *datetime* "Data da Coleta". 
* Saída: *DataFrame* com as quinzenas do mês.

**<font color="#000099">cria_semana:</font>**

* Entrada: *DataFrame* com coluna do tipo *datetime* "Data da Coleta". 
* Saída: *DataFrame* com as semanas do mês.

```python
# Funções para criar quinzenas e semanas

def cria_quinzenas(dataframe):
    primeiro = (f"{dataframe['Data da Coleta'][0].year}-{dataframe['Data da Coleta'][0].month}-01")
    primeiro=pd.to_datetime(primeiro, format = "%Y/%m/%d")
    ultimo =(f"{dataframe['Data da Coleta'][len(dataframe)-1].year}-{dataframe['Data da Coleta'][len(dataframe)-1].month}-{dataframe['Data da Coleta'][len(dataframe)-1].day}")
    ultimo=pd.to_datetime(ultimo, format = "%Y/%m/%d")

    dic = {}
    total_dias = ultimo-primeiro
    quinzenas = total_dias/15
    dia=primeiro
    for i in range(1,quinzenas.days+1):
        dia_calc = (dia+timedelta(days=15))
        dic[i]=(dia_calc)
        dia = dia_calc
    dic[i+1]=ultimo
    
    dia=primeiro
    saida = pd.DataFrame()
    for i in dic.items():
        while dia <=i[1]:
            a = pd.DataFrame(data=[i[0]],index=[dia],columns=['Quinzena'])
            saida = pd.concat([saida,a])
            dia = dia + timedelta(days=1)
    df = pd.merge(right= dataframe,left = saida,how='right',right_on = ['Data da Coleta'],left_index=True)
    return df

def cria_semana(dataframe):
    primeiro = (f"{dataframe['Data da Coleta'][0].year}-{dataframe['Data da Coleta'][0].month}-01")
    primeiro=pd.to_datetime(primeiro, format = "%Y/%m/%d")
    ultimo =(f"{dataframe['Data da Coleta'][len(dataframe)-1].year}-{dataframe['Data da Coleta'][len(dataframe)-1].month}-{dataframe['Data da Coleta'][len(dataframe)-1].day}")
    ultimo=pd.to_datetime(ultimo, format = "%Y/%m/%d")

    dic = {}
    total_dias = ultimo-primeiro
    semanas = total_dias/7
    dia=primeiro
    for i in range(1,semanas.days+1):
        dia_calc = (dia+timedelta(days=7))
        dic[i]=(dia_calc)
        dia = dia_calc
    dic[i+1]=ultimo
    
    dia=primeiro
    saida = pd.DataFrame()
    for i in dic.items():
        while dia <=i[1]:
            a = pd.DataFrame(data=[i[0]],index=[dia],columns=['Semana'])
            saida = pd.concat([saida,a])
            dia = dia + timedelta(days=1)
    df = pd.merge(right= dataframe,left = saida,how='right',right_on = ['Data da Coleta'],left_index=True)
    return df
```

<h3><b>Diante disso, tentamos criar as correlações entre as variantes do preço de venda, sendo elas: "Preço Mínimo", "Preço Médio" e "Preço Máximo" para saber se o fator regional impactava nos resultados</b></h3>

```python
# Para isso criamos um novo DataFrame com os valores de mínimo, médio e máximo:

maximas_por_região = dados_jun_jul.groupby(['Produto','Região'])[['Valor de Venda']].max()
maximas_por_região.columns = ['Máxima']
minimas_por_região = dados_jun_jul.groupby(['Produto','Região'])[['Valor de Venda']].min()
minimas_por_região.columns = ['Minima']
media_geral = pd.concat([media_por_regiao,maximas_por_região,minimas_por_região],axis=1)
media_geral.rename(columns={'Valor de Venda': 'Média'}).round(3)
```

```python
# Plotando um gráfico de calor:

plt.figure(figsize=(10,5))
sns.heatmap(media_geral.corr(), vmin=-1.0, vmax=1.0, cmap='coolwarm', annot=True)
print()
```

<h3><b>O gráfico de calor acima indica uma forte correlação entre as variantes de preço, porém, não destaca a intensidade dessa correlação em função da região. Motivo pelo qual, testamos a existência de correlação através da Transposta do DataFrame</b></h3>

```python
# Aplicando a correlação com a visualização do Dataframe de forma Transposta:

plt.figure(figsize=(15,10))
sns.heatmap(media_geral.T.corr(), vmin=-1.0, vmax=1.0, cmap='coolwarm', annot=True)
print()
```

<h3><b>O Gráfico de calor resultante da Transposta do Dataframe de Variação de Preços, reitera a forte correlação entre as variáveis, independente da região amostrada, evidenciando com isso não ser um recurso válido para tomada de decisões nessa situação.</b></h3>

<h3><b>Carregamos os dados do Ibov, Petróleo e Dolár:</b></h3>

```python
ibov = yf.download("^BVSP")
brent = yf.download("BZ=F")
dolar = yf.download("BRL=X")
```

```python
inflacao = pd.read_excel('../input/inflacao-2021/inflacao.xlsx')
inflacao_acumulada = inflacao[['Mês','Acum. Ano%']]
inflacao_acumulada.set_index('Mês',inplace = True)
```

```python
brent.drop(['Open','High','Low','Adj Close','Volume'],axis=1,inplace=True)
ibov.drop(['Open','High','Low','Adj Close','Volume'],axis=1,inplace=True)
dolar.drop(['Open','High','Low','Adj Close','Volume'],axis=1,inplace=True)
```

```python
# Unindo os ativos com os dados de junho e julho
dados_merge = pd.merge(right=dados_ano_2021,left=brent,how='right',right_on='Data da Coleta',left_index=True)
dados_merge = pd.merge(right=dados_merge,left=ibov,how='right',right_on='Data da Coleta',left_index=True)
dados_merge = pd.merge(right=dados_merge,left=dolar,how='right',right_on='Data da Coleta',left_index=True)
dados_merge = pd.merge(right=dados_merge,left=inflacao_acumulada,how='right',right_on='Data da Coleta',left_index=True)
dados_merge.rename(columns={'Close':'Dolar','Close_x':'Ibov','Close_y':'Brent','Acum. Ano%':'Inflacao'},inplace=True)
dados_merge = dados_merge.groupby('Data da Coleta').mean()
```

```python
dados_merge.fillna(method='bfill',inplace=True)
dados_merge.fillna(0,inplace=True)
```

```python
corr_merge = dados_merge[['Inflacao','Dolar','Ibov','Brent','Valor de Venda']].corr()
plt.figure(figsize=(7,4))
sns.heatmap(corr_merge,vmin =-1.0,vmax=1.0,cmap='RdBu',annot=True)
print()
```
