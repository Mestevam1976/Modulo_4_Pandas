### 1 - Como se comportaram o preço dos combustíveis durante os dois meses citados? Os valores do etanol e da gasolina tiveram uma tendência de queda ou diminuição?

```python
# TENDÊNCIA DO ETANOL NO PERÍODO DE JUNO A JULHO - ATRAVÉS DO DESCRIBE:

tendencia_etanol = df_merged.loc[(df_merged['Produto']=='ETANOL')&(df_merged['Valor de Venda'])]

tendencia_etanol['Valor de Venda'].describe()
```

### R: Os preços se elevaram no período de junho a julho de 2021.

```python
# TENDÊNCIA DO ETANOL SOMENTE MÊS DE JUNHO DE 2021 - ATRAVÉS DA MÉDIA DIRETA:

df_june['Valor de Venda'] = df_june['Valor de Venda'].str.replace(',','.').astype(float)

preco_medio_junho = df_june.loc[(df_june['Produto']=='ETANOL')&(df_june['Valor de Venda'])]

preco_m_junho = preco_medio_junho['Valor de Venda'].mean()

preco_m_junho
```

```python
# TENDÊNCIA DO ETANOL SOMENTE MÊS DE JULHO DE 2021 - ATRAVÉS DA MÉDIA DIRETA:

df_july['Valor de Venda'] = df_july['Valor de Venda'].str.replace(',','.').astype(float)

preco_medio_julho = df_july.loc[(df_july['Produto']=='ETANOL')&(df_july['Valor de Venda'])]

preco_m_julho = preco_medio_julho['Valor de Venda'].mean()

preco_m_julho
```

```python
plt.figure(figsize=(10,6))
plt.title('Evolução do Preço Médio de Venda do Etanol no período de Junho a Julho de 2021')
x1 = 'Junho','Julho'
y1 = preco_m_junho, preco_m_julho 
plt.grid() 
plt.plot(x1,y1)
```

```python
# TENDÊNCIA DA GASOLINA NO PERÍODO DE JUNO A JULHO - ATRAVÉS DO DESCRIBE:

tendencia_gasolina = df_merged.loc[(df_merged['Produto']=='GASOLINA')&(df_merged['Valor de Venda'])]

tendencia_gasolina['Valor de Venda'].describe()
```

### 4 -  Qual o preço médio da gasolina e do etanol por estado?

```python
## PLOTANDO UM GRÁFICO DE LINHA COM VARIAÇÃO DO PREÇO MÉDIO DO ETANOL POR ESTADO - PERÍODO DE JUNHO A JULHO DE 2021

plt.figure(figsize=(15,6))
plt.title('Preço Médio do Etanol por Estado - Período de Junho a Julho de 2021')
plt.grid()
plt.xlabel('Estados')
plt.ylabel('Valores em Reais')
plt.plot(preco_medio_etanol_estado, label='Preço Médio por Estado')

## CRIANDO A LINHA DO PREÇO MÉDIO:
preco_medio =[]
n = 27 # Número de Estados da Federação
contador = 0
while contador < n:
    for i in df_preco_medio[:1].values: # Selecionando apenas o Etanol 
        preco_medio.append(i)
        contador +=1
plt.plot(preco_medio, label='Preço Médio Nacional')
plt.legend()
```

```python
## PREÇO MÉDIO DA GASOLINA POR ESTADO NO PERÍODO DE JUNHO A JULHO DE 2021:

df_auxiliar_10 = df_merged[['Valor de Venda', 'Produto', 'Estado']]

df_auxiliar_11 = df_auxiliar_10.loc[(df_auxiliar_8['Produto']=='GASOLINA')& (df_auxiliar_8['Valor de Venda'])& (df_auxiliar_8['Estado'])]

df_preco_medio_geral = df_auxiliar_11.groupby(['Estado']).mean()

preco_medio_gasolina_estado = df_preco_medio_geral.rename(columns = {'Valor de Venda': 'Preço Médio'}).round(2) # .round(2) para definir duas casas decimais 

preco_medio_gasolina_estado
```

```python
## PLOTANDO UM GRÁFICO DE LINHA COM VARIAÇÃO DO PREÇO MÉDIO DA GASOLINA POR ESTADO - PERÍODO DE JUNHO A JULHO DE 2021

plt.figure(figsize=(15,6))
plt.title('Preço Médio da Gasolina por Estado - Período de Junho a Julho de 2021')
plt.grid()
plt.xlabel('Estados')
plt.ylabel('Valores em Reais')
plt.plot(preco_medio_gasolina_estado, label='Preço Médio por Estado')

## CRIANDO A LINHA DO PREÇO MÉDIO:
preco_medio =[]
n = 27 # Número de Estados da Federação
contador = 0
while contador < n:
    for i in df_preco_medio[1:2].values: # Selecionando apenas a Gasolina 
        preco_medio.append(i)
        contador +=1
plt.plot(preco_medio, label='Preço Médio Nacional')
plt.legend()
```

```python
## PREÇO MÉDIO DA GASOLINA ADITIVADA POR ESTADO NO PERÍODO DE JUNHO A JULHO DE 2021:

df_auxiliar_12 = df_merged[['Valor de Venda', 'Produto', 'Estado']]

df_auxiliar_13 = df_auxiliar_12.loc[(df_auxiliar_8['Produto']=='GASOLINA ADITIVADA')& (df_auxiliar_8['Valor de Venda'])& (df_auxiliar_8['Estado'])]

df_preco_medio_geral = df_auxiliar_13.groupby(['Estado']).mean()

preco_medio_gasolina_adt_estado = df_preco_medio_geral.rename(columns = {'Valor de Venda': 'Preço Médio'}).round(2) # .round(2) para definir duas casas decimais 

preco_medio_gasolina_adt_estado
```

```python
## PLOTANDO UM GRÁFICO DE LINHA COM VARIAÇÃO DO PREÇO MÉDIO DA GASOLINA ADITIVADA POR ESTADO - PERÍODO DE JUNHO A JULHO DE 2021

plt.figure(figsize=(15,6))
plt.title('Preço Médio da Gasolina Aditivada por Estado - Período de Junho a Julho de 2021')
plt.grid()
plt.xlabel('Estados')
plt.ylabel('Valores em Reais')
plt.plot(preco_medio_gasolina_adt_estado, label='Preço Médio por Estado')

## CRIANDO A LINHA DO PREÇO MÉDIO:
preco_medio =[]
n = 27 # Número de Estados da Federação
contador = 0
while contador < n:
    for i in df_preco_medio[2:].values: # Selecionando apenas a Gasolina Aditivada 
        preco_medio.append(i)
        contador +=1
plt.plot(preco_medio, label='Preço Médio Nacional')
plt.legend()
```

```python
## PREÇO MÉDIO DOS TRÊS TIPOS DE COMBUSTÍVEIS POR ESTADO NO PERÍODO DE JUNHO A JULHO DE 2021:

df_auxiliar_22 = df_merged[['Valor de Venda', 'Produto', 'Estado']]

#df_auxiliar_23 = df_auxiliar_22.loc[(df_auxiliar_8['Produto'])& (df_auxiliar_8['Valor de Venda'])& (df_auxiliar_8['Estado'])]

df_preco_medio_geral = df_auxiliar_22.groupby(['Estado', 'Produto']).mean()

preco_medio_combustiveis_estado = df_preco_medio_geral.rename(columns = {'Valor de Venda': 'Preço Médio'}).round(2) # .round(2) para definir duas casas decimais 

preco_medio_combustiveis_estado
```

```python
## PLOTANDO UM GRÁFICO DE LINHA COM VARIAÇÃO DO PREÇO MÉDIO DOS COMBUSTÍVEIS POR ESTADO - PERÍODO DE JUNHO A JULHO DE 2021

x = df_merged['Estado'].unique()

y1 = preco_medio_etanol_estado['Preço Médio'].values

y2 = preco_medio_gasolina_estado['Preço Médio'].values

y3 = preco_medio_gasolina_adt_estado['Preço Médio'].values


plt.figure(figsize=(15,8))
plt.title('Preço Médio dos Combustíveis por Estado - Período de Junho a Julho de 2021')
plt.grid()
plt.xlabel('Estados')
plt.ylabel('Valores em Reais')
plt.plot(x, y1, label='Etanol')
plt.plot(x, y2, label='Gasolina')
plt.plot(x, y3, label='Gasolina Aditivada')
plt.legend()
```

### 5 - Qual o município que possui o menor preço para a gasolina e para o etanol?

```python
## CIDADE COM O MENOR VALOR PRATICADO PARA O COMBUSTÍVEL ETANOL

df_auxiliar_14 = df_merged.groupby(['Produto', 'Município'])

df_auxiliar_15 = df_auxiliar_14[["Valor de Venda"]].min()

etanol_min_municipio = df_auxiliar_15.loc['ETANOL'].sort_values(by='Valor de Venda')

etanol_min_municipio.head(1)
```

```python
## CIDADE COM O MENOR VALOR PRATICADO PARA O COMBUSTÍVEL GASOLINA

df_auxiliar_16 = df_merged.groupby(['Produto', 'Município'])

df_auxiliar_17 = df_auxiliar_16[["Valor de Venda"]].min()

gasolina_min_municipio = df_auxiliar_17.loc['GASOLINA'].sort_values(by='Valor de Venda')

gasolina_min_municipio.head(1)
```

```python
## CIDADE COM O MENOR VALOR PRATICADO PARA O COMBUSTÍVEL GASOLINA ADITIVADA

df_auxiliar_18 = df_merged.groupby(['Produto', 'Município'])

df_auxiliar_19 = df_auxiliar_18[["Valor de Venda"]].min()

gasolina_adt_min_municipio = df_auxiliar_19.loc['GASOLINA ADITIVADA'].sort_values(by='Valor de Venda')

gasolina_adt_min_municipio.head(1)
```

```python
## COLETANDO AS CIDADES COM OS MENORES VALORES PRATICADOS:

cidades = []
cidades.append(etanol_min_municipio.index[0])
cidades.append(gasolina_min_municipio.index[0])
cidades.append(gasolina_adt_min_municipio.index[0])

## CRIANDO AS MENSAGENS COM VALORES DAS CIDADES SELECIONADAS:

mensagem_1 = etanol_min_municipio['Valor de Venda'].min()
mensagem_2 = gasolina_min_municipio["Valor de Venda"].min()
mensagem_3 = gasolina_adt_min_municipio["Valor de Venda"].min()
```

```python
## CRIANDO UM MAPA DO BRASIL ATRAVÉS DO FOLIUM:

brasil = folium.Map(
    location=[-16.1237611, -59.9219642],    # Coordenadas retiradas do Google Maps
    zoom_start=4   # Define o nível de zoom do mapa no carregamento, o default é 10
)

## PEGANDO AS COORDENADAS GEOGRÁFICAS (LATITUDE E LONGITUDE) POR NOME DE CIDADE ATRAVÉS DA BIBLIOTECA GEOPY:

from geopy.geocoders import Nominatim 

a= []

for i in cidades:
    loc = Nominatim(user_agent="GetLoc") 
    getLoc = loc.geocode(i)
    location = getLoc
    a.append([location.latitude, location.longitude])
    
coordenadas_1 = a[0]
coordenadas_2 = a[1]
coordenadas_3 = a[2]

# INSERINDO MARCADORES NO MAPA DO BRASIL COM BASE NAS COORDENADAS GEOGRÁFICAS:

mapa = brasil

folium.Marker(
    location=[coordenadas_1[0], coordenadas_1[1]],
    popup=mensagem_1,
    icon=folium.Icon(color="red", icon="info-sign"),
).add_to(mapa)

folium.Marker(
    location=[coordenadas_2[0], coordenadas_2[1]],
    popup=mensagem_2,
    icon=folium.Icon(color="green", icon="info-sign"),
).add_to(mapa)

folium.Marker(
    location=[coordenadas_3[0], coordenadas_3[1]],
    popup=mensagem_3,
    icon=folium.Icon(color="blue", icon="info-sign"),
).add_to(mapa)


mapa
```

### 7- Qual a região que possui o maior valor médio da gasolina?

```python
df_auxiliar_18 = df_merged[['Produto', 'Região', 'Valor de Venda']]

df_auxiliar_19 = df_auxiliar_18.loc[(df_auxiliar_18['Produto']=='GASOLINA') & (df_auxiliar_18['Região']) & (df_auxiliar_18['Valor de Venda'])]

df_gasolina_preco_medio_geral = df_auxiliar_19.groupby(['Produto','Região']).mean().round(2)

df_gasolina_preco_medio_geral.sort_values(by="Valor de Venda", ascending=False).head(1)
```

### 9 - Há alguma correlação entre o valor do combustível (gasolina e etanol) e a região onde ele é vendido?

```python
## PREÇO MÉDIO DOS COMBUSTÍVEIS POR REGIÃO NO PERÍODO DE JUNHO A JULHO DE 2021.

df_auxiliar_1 = df_merged[['Valor de Venda', 'Produto', 'Região']]

df_preco_medio_geral = df_auxiliar_1.groupby(['Produto', 'Região']).mean()

df_preco_medio_geral.rename(columns = {'Valor de Venda': 'Preço Médio'}).round(2) # .round(2) para definir duas casas decimais 
```

```python
df_auxiliar_1 = df_merged[['Valor de Venda', 'Produto', 'Região']]


df_auxiliar_2 = df_auxiliar_1.groupby(['Produto', 'Região']).max()

df_auxiliar_2
```

```python
df_auxiliar_3 = df_merged[['Valor de Venda', 'Produto', 'Região']]


df_auxiliar_3 = df_auxiliar_1.groupby(['Produto', 'Região']).min()

df_auxiliar_3
```

```python
## CRIANDO NOVO DATAFRAME COM OS VALORES MÍNIMO, MÉDIO E MÁXIMO:

df_preco_medio_geral['Preço máximo'] = df_auxiliar_2['Valor de Venda']

df_preco_medio_geral['Preço mínimo'] = df_auxiliar_3['Valor de Venda']

df_preco_medio_geral
```

```python
## APLICANDO O TESTE DE CORRELAÇÃO COM O DATAFRAME NORMAL

df_preco_medio_geral.corr()
```
