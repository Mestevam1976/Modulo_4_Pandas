# Dados de Junho 2021

```python
import pandas as pd
import numpy as np

data_june = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-06-gasolina-etanol.csv', sep=';')

df_june = pd.DataFrame(data_june)

df_june
```

## Checando os combustíveis diferentes:

```python
df_june['Produto'].unique()
```

## Checando os valores praticados na região Sudeste do combustível tipo Gasolina comum

```python
df_re_se_gas = df_june.loc[(df_june['Valor de Venda']!='NaN') & (df_june['Região - Sigla'] == 'SE') & (df_june['Produto'] == 'GASOLINA')]

df_re_se_gas['Valor de Venda'].describe()
```

## Checando os valores praticados na região Sudeste do combustível tipo Gasolina Aditivada


```python
df_re_se_gas = df_june.loc[(df_june['Valor de Venda']!='NaN') & (df_june['Região - Sigla'] == 'SE') & (df_june['Produto'] == 'GASOLINA ADITIVADA')]

df_re_se_gas['Valor de Venda'].describe()
```

## Checando os valores praticados na região Sudeste do combustível tipo Etanol


```python
df_re_se_gas = df_june.loc[(df_june['Valor de Venda']!='NaN') & (df_june['Região - Sigla'] == 'SE') & (df_june['Produto'] == 'ETANOL')]

df_re_se_gas['Valor de Venda'].describe()
```

```python
df_re_se_gas = df_june.loc[(df_june['Valor de Venda']!='NaN')  & (df_june['Produto'] == 'ETANOL')]

#& (df_june['Região - Sigla'] == 'SE')

df_re_se_gas

d1 = df_re_se_gas['Valor de Venda'].str.replace(',','.').astype(float) # Convertendo string para float

d1.describe()
```
