# Dados de Julho 2021

```python
import pandas as pd
import numpy as np

data_july = pd.read_csv('https://www.gov.br/anp/pt-br/centrais-de-conteudo/dados-abertos/arquivos/shpc/dsan/2021/2021-07-gasolina-etanol.csv', sep=';')

df_july = pd.DataFrame(data_july)

df_july
```

## Checando os combustíveis diferentes:

```python
df_july['Produto'].unique()
```

## Checando os valores praticados na região Sudeste do combustível tipo Gasolina comum

```python
df_sudeste_gas = df_july.loc[(df_july['Valor de Venda']!='NaN') & (df_july['Região - Sigla'] == 'SE') & (df_july['Produto'] == 'GASOLINA')]

df_sudeste_gas['Valor de Venda'].describe()
```

## Checando os valores praticados na região Sudeste do combustível tipo Gasolina Aditivada

```python
df_sudeste_gasAd = df_july.loc[(df_july['Valor de Venda']!='NaN') & (df_july['Região - Sigla'] == 'SE') & (df_july['Produto'] == 'GASOLINA ADITIVADA')]

df_sudeste_gasAd['Valor de Venda'].describe()
```

## Checando os valores praticados na região Sudeste do combustível tipo Etanol

```python
df_sudeste_etanol = df_july.loc[(df_july['Valor de Venda']!='NaN') & (df_july['Região - Sigla'] == 'SE') & (df_july['Produto'] == 'ETANOL')]

df_sudeste_etanol['Valor de Venda'].describe()
```

## Observado que os valores para serem contabilizados em dados estatísticos deverão ser tratados: substituindo virgulas por pontos na separação dos decimais.


```python
d1 = df_sudeste_etanol['Valor de Venda'].str.replace(',','.').astype(float)

d1.describe()
```
